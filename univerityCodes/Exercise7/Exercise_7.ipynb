### Task1

```python
#  solve and plot the results of the differential equation : 
#y'' + y'+y = 0; y(0) =1 ; y'(0) = 0
import numpy as np
import matplotlib.pyplot as plt
from scipy import *
from scipy.integrate import odeint, ode
def dw(w,t):
    y, p = w[0],w[1]
    dy = p
    dp = -(p+y)
    return [dy,dp]
w0 = [1.0,0.0]
t = np.linspace(0,15,1000)
w1 = odeint(dw,w0,t)
plt.plot(t,w1[:,0])
```

### Task2

```python
# Write a script that creates different polynomial fits 
#by varying the order by means of a for loop
import numpy as np
import matplotlib.pyplot as plt
 
x = np.array([0,1,2,3,4,5])
y = np.array([15,10,9,6,2,0])

def fitting_function(x,b): #a multiple b power c offset
    return -3*x**b+15

plt.plot(x,y,'o',label="data")
arr = np.arange(0.7,1.5,0.1)
for a in arr:
    plt.plot(x,fitting_function(x,a), label = "power = " + str(a))
    plt.legend()
```

### Task 3

```python
#Use SymPy to solve the differential equation of task 1
from sympy import *
t = Symbol('t')
y = Function("y") # y'' + y' +y = 0
w = dsolve(Derivative(y(t),t,t) + Derivative(y(t),t) + y(t),y(t))
C2 = solve(w.subs(t,0).subs(y(0),1))[0]
w = w.subs("C2",C2)
w = w.subs("C1",0)
sol = np.zeros(1000)
T = np.linspace(0,15,1000)
for k in range(0,1000):
    sol[k] = w.subs(t,T[k]).rhs
```

```python
plt.plot(T,sol,label = "sympy")
plt.plot(T,w1[:,0], label = "scipy")
plt.legend()
```
