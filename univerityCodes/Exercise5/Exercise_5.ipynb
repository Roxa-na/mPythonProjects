```python
# Write a script that evaluates a photo (sky.png) of a section of
#the sky and uses it to determine the degree of cloudiness.
img = plt.imread("sky.png")
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
img_blue = img[:,:,2]
```

```python
plt.imshow(img_blue)
plt.colorbar()
```

```python
img_gray = img.mean(axis = 2)
```

```python
plt.imshow(img_gray)
```

```python
img_norm = img_blue/img_gray
```

```python
plt.imshow(img_norm, cmap = plt.cm.gray) #answer first picture
plt.colorbar()
```

```python
mask = img_norm > 1.3
```

```python
plt.imshow(mask) # answer second picture
plt.colorbar()
```

```python
count_blue = np.count_nonzero(mask)
```

```python
count_blue
```

```python
mask.size
```

```python
blueDegree = (count_blue/mask.size)
```

```python
blueDegree
```
