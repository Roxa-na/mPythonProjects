### Task 1:

```python
# money format N15000723228 
# one letter 11 numbers: 10 code digit and check digit
#Write a program that accepts a Euro serial number (as user input) and
#checks if the serial number is correct.
# the checking is done as the following
# Take the position of the letter in the alphabet instead of the country code (starting with 1 for A)
# Make the checksum from all digits (except the check digit). (summing all digits)
#Calculate and take the rest of the division by nine (9).
#Subrtract the calculated remainder from eight (8).
#  this difference (8 - remainder) is equal to the check digit. if so correct.
def check_serial_number(s):
    country_code = ord(s[0]) - 64
    new_serial_number = str(country_code) + s[1:11]
    check_digit = int(s[11])
    list_string_number = list(new_serial_number)
    list_int_number = [int(item) for item in list_string_number]
    Sum = sum(list_int_number)
    remainder = Sum % 9
    if (8 - remainder == check_digit):
        return "correct"
    else:
        return "incorrect"
```

```python
check_serial_number("N15000723228")
```

```python
check_serial_number("U61068861284")
```

```python
check_serial_number("V23197584928")
```

```python
#Write a function “median” that calculates the median of a set of numbers.
def median(lis):
    lis.sort()
    length = len(lis)
    mid = int(length/2)
    if length%2 == 0:
       med = (lis[mid-1] + lis[mid])/2
    else:
        med = lis[int((length + 1)/2)-1] 
    return med
```
