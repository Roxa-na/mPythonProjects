### Task 2

```python
#The file "sunspots.txt" contains the number of sunspots observed per month since January 1749
# Write a Python script that reads the data and plots the number of sunspots per month over
#time. 
# Modify the script to show only the first 1000 data points
# Write a function move_avg (Data, Radius) that calculates the moving average 

filename = "sunspots.txt"
data = genfromtxt(filename)
for m in range (12):
    n = m
    while(n < 3143):
        data[n,0] = m
        n = n+12

p= [0]*12
for x in range(12):
    for d in (data):
        if (d[0] == x):
            p[x] = p[x] + d[1]
month = [i for i in range(12)]
fig, ax = plt.subplots()
ax.bar(month, p)
ax.set_title("number of sunspots")
ax.set_xlabel("Month")
ax.set_ylabel("Monthly sunspots");
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 2]) # main axes these are in percent
axes2 = fig.add_axes([0.2,1, 0.4, 0.3]) # inset axes

y = data[:1000]
p1= [0]*12
for x in range(12):
    for d in (y):
        if (d[0] == x):
            p1[x] = p1[x] + d[1]
# main figure
axes1.bar(month, p1)
axes1.set_xlabel('month')
axes1.set_ylabel('number of sunspots')
axes1.set_title('monthly sunspots for 1000')

# insert
axes2.bar(month, p)
axes2.set_title("number of sunspots")
axes2.set_xlabel("Month")
axes2.set_ylabel("Monthly sunspots");
```

```python
def move_avg(data,r):
    q = 1/(2*r)
    yk = [0]*12
    for ind,d in enumerate(data):
        s = 0
        for m in range(-r,r):
            s = s + r + d;
        yk[ind] = s/q
    return yk
move_avg(p,50)
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12,5))

axes[0].bar(month,move_avg(p,50))
axes[0].set_title("number of sunspots")
axes[0].set_xlabel("Month")
axes[0].set_ylabel("Monthly sunspots");
axes[1].bar(month,p)
axes[1].set_title("average of sunspots")
axes[1].set_xlabel("Month")
axes[1].set_ylabel("average of Monthly sunspots");
fig.tight_layout()
     
```

```python
#Find the number of π by randomly distributing points in a
#coordinate system of length 1 and height 1.
# Use the ratio of the number of points in the
#circle to the total number of points to determine the number π.
n = 20000
ins = []
x = np.random.random(n)
y = np.random.random(n)
for i in range(n):
    if (np.sqrt(x[i]**2+y[i]**2) < 1):
        ins.append((x,y))
```

```python
4 * len(ins)/float(n) # pi
```
